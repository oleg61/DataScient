```python
sbr.distplot(df_1['distance_km'])
sbr.distplot(df_2['distance_km'])
plt.title('Динамика распределения дистаницй поездок')
plt.legend(['Economy', 'Comfort'])
plt.show()
```

```python
# Подключаем библиотеки статистических критериев и зададим уровень значимости
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
#Уровень значимости
alpha = 0.05
```
